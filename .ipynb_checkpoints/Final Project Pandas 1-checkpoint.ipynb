```python
# WRONG DATE FORMAT:
OilDates = OilDF["Date"]
OilDF.head()
```

```python
# THIS CELL "FIXES" THE STRING DATE OF THE OIL CSV TO BE MORE LIKE THE FRED DATE FORMAT:

day = ""
month =""
MonthNums = {'Jan':"01", 'Feb':"02",'Mar':"03", 'Apr':"04",'May':"05", 'Jun':"06",'Jul':"07",
             'Aug':"08", 'Sep':"09", 'Oct':"10",'Nov':"11", 'Dec':"12"}

FixedDates=[]
count=0
for ind,row in OilDF.iterrows():
    #print(row["Date"])
    unfixedYear= row["Date"][-2:]
    if(int(unfixedYear)<21):
        FixedYear = "20"+unfixedYear
        #print(FixedYear)
    else:
        FixedYear = "19"+unfixedYear
        #print(FixedYear)
    if(row["Date"][2]=="-"):
#         print("this has two digits: " +row["Date"][0:2])
        day = (row["Date"][0:2])
#         print("Month: " + row["Date"][3:6])
        month = (MonthNums[row["Date"][3:6]])
    elif (row["Date"][1]=="-"):
#         print("this has one digit: " +row["Date"][0:1])
        day = "0"+ (row["Date"][0:1])
#         print("month: "+ row["Date"][2:5])
        month = (MonthNums[row["Date"][2:5]])
    FixedDate = FixedYear +"-"+month+"-"+day
    FixedDates.append(FixedDate)
    count +=1
    print(str(count) +":"+ FixedDate)

#print(FixedDates)
```

```python
OilDF["DATE"]= FixedDates
OilDF
```
